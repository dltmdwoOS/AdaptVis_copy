```python
import pandas as pd
import json

# 1. JSON 파일 로드 (파일 경로에 맞게 수정 필요)
# data = pd.read_json('your_file_path.json') 

# (예시 데이터 생성 - 실제 사용 시 이 부분은 제외하고 위 read_json을 사용하세요)
with open("output/results1.5_Controlled_Images_A_adapt_vis_1.0_0.4_fouroption_True.json", 'r') as f:
    data = json.load(f)
    
df = pd.DataFrame(data)

# 2. 제외할 단답형 키워드 정의 (소문자로 통일)
target_keywords = {'right', 'left', 'on', 'under'}

# 3. 필터링 로직 적용
# (1) 공백 제거(.str.strip) -> (2) 소문자 변환(.str.lower) -> (3) 키워드 목록에 없는 경우(~.isin)만 선택
filtered_df = df[~df['Generation'].str.strip().str.lower().isin(target_keywords)].copy()

# 4. 결과 확인 및 저장
print(f"총 데이터 개수: {len(df)}")
print(f"필터링된 문장형 데이터 개수: {len(filtered_df)}")
print(filtered_df[['Generation', 'Golden']])

# 필요시 CSV로 저장
# filtered_df.to_csv('sentence_cases.csv', index=False, encoding='utf-8-sig')
```

```python
def counter(x):
    c = 0
    rels = ['left', 'right', 'on', 'under']
    for rel in rels:
        if rel in x.lower():
            c += 1
    return c

filtered_df['count'] = filtered_df['Generation'].apply(counter)
```

```python
filtered_df[filtered_df["count"] >= 2]
```

```python
def check_floor(x):
    if 'floor' in x.lower():
        return True
    return False

filtered_df['contains_floor'] = filtered_df['Generation'].apply(check_floor)
filtered_df[filtered_df["contains_floor"] == True]
```

```python
import json
import re

def create_relational_datasets(input_file_path, output_on_path, output_under_path):
    """
    JSONL 파일을 읽어 'On' 관계와 'Under' 관계를 묻는 두 개의 데이터셋을 생성합니다.
    """
    
    # 질문 구조에서 (is/are), (obj1), (obj2)를 추출하기 위한 정규표현식
    # 예: Where is the beer bottle in relation to the armchair?
    # group(1): is/are
    # group(2): obj1 (smaller one)
    # group(3): obj2 (bigger one)
    pattern = r"Where (is|are) the (.+?) in relation to the (.+?)\?"
    
    on_dataset = []
    under_dataset = []
    
    try:
        with open(input_file_path, 'r', encoding='utf-8') as f:
            for line in f:
                entry = json.loads(line)
                
                question_text = entry.get('question', '')
                match = re.search(pattern, question_text)
                
                if match:
                    be_verb = match.group(1)      # is 또는 are
                    obj1 = match.group(2)         # 객체 1
                    obj2 = match.group(3)         # 객체 2
                    
                    # 원본 답변 추출 (대소문자 통일)
                    original_answer = entry['answer'][0].lower()
                    
                    # 문장 시작이므로 Be동사 첫 글자 대문자화
                    be_verb_cap = be_verb.capitalize()
                    
                    # -------------------------------------------------------
                    # 1. On Dataset 생성 로직
                    # -------------------------------------------------------
                    one_shot = "USER: Is the calculator on the desk? Answer with yes or no.\nASSISTANT: Think step by step. In this picture, the gray calculator is placed on top of the wooden desk, so the answer is yes.\n"
                    new_question_on = one_shot + f"<image>\nUSER: {be_verb_cap} the {obj1} on the {obj2}? Answer with yes or no.\nASSISTANT: Think step by step. "
                    
                    # 원본이 "on"일 때만 Yes, 나머지는 No
                    new_answer_on = "Yes" if original_answer == "on" else "No"
                    
                    on_dataset.append({
                        "id": entry['id'],
                        "question": new_question_on,
                        "answer": [new_answer_on]
                    })
                    
                    # -------------------------------------------------------
                    # 2. Under Dataset 생성 로직
                    # -------------------------------------------------------
                    new_question_under = f"<image>\nUSER: {be_verb_cap} the {obj1} under the {obj2}? Answer with yes or no.\nASSISTANT:"
                    
                    # 원본이 "under"일 때만 Yes, 나머지는 No
                    new_answer_under = "Yes" if original_answer == "under" else "No"
                    
                    under_dataset.append({
                        "id": entry['id'],
                        "question": new_question_under,
                        "answer": [new_answer_under]
                    })

        # 결과 저장
        with open(output_on_path, 'w', encoding='utf-8') as f_on:
            for item in on_dataset:
                f_on.write(json.dumps(item) + '\n')
                
        with open(output_under_path, 'w', encoding='utf-8') as f_under:
            for item in under_dataset:
                f_under.write(json.dumps(item) + '\n')
                
        print(f"성공적으로 처리되었습니다.")
        print(f"- On 데이터셋 저장됨: {output_on_path} (총 {len(on_dataset)}개)")
        print(f"- Under 데이터셋 저장됨: {output_under_path} (총 {len(under_dataset)}개)")

    except FileNotFoundError:
        print(f"에러: 입력 파일 '{input_file_path}'을 찾을 수 없습니다.")
    except Exception as e:
        print(f"처리 중 에러가 발생했습니다: {e}")

# 사용 예시
input_jsonl = "prompts/Controlled_Images_A_with_answer_four_options.jsonl"          # 원본 파일 경로
output_on_jsonl = "prompts/Controlled_Images_A_On_with_answer_two_options.jsonl"   # 결과 파일 경로 (On)
output_under_jsonl = "prompts/Controlled_Images_A_Under_with_answer_two_options.jsonl" # 결과 파일 경로 (Under)

create_relational_datasets(input_jsonl, output_on_jsonl, output_under_jsonl)
    
```
