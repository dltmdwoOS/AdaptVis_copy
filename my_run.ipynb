```python
import pandas as pd
import json

# 1. JSON 파일 로드 (파일 경로에 맞게 수정 필요)
# data = pd.read_json('your_file_path.json') 

# (예시 데이터 생성 - 실제 사용 시 이 부분은 제외하고 위 read_json을 사용하세요)
with open("output/results1.5_Controlled_Images_A_adapt_vis_1.0_0.4_fouroption_True.json", 'r') as f:
    data = json.load(f)
    
df = pd.DataFrame(data)

# 2. 제외할 단답형 키워드 정의 (소문자로 통일)
target_keywords = {'right', 'left', 'on', 'under'}

# 3. 필터링 로직 적용
# (1) 공백 제거(.str.strip) -> (2) 소문자 변환(.str.lower) -> (3) 키워드 목록에 없는 경우(~.isin)만 선택
filtered_df = df[~df['Generation'].str.strip().str.lower().isin(target_keywords)].copy()

# 4. 결과 확인 및 저장
print(f"총 데이터 개수: {len(df)}")
print(f"필터링된 문장형 데이터 개수: {len(filtered_df)}")
print(filtered_df[['Generation', 'Golden']])

# 필요시 CSV로 저장
# filtered_df.to_csv('sentence_cases.csv', index=False, encoding='utf-8-sig')
```

```python
def counter(x):
    c = 0
    rels = ['left', 'right', 'on', 'under']
    for rel in rels:
        if rel in x.lower():
            c += 1
    return c

filtered_df['count'] = filtered_df['Generation'].apply(counter)
```

```python
filtered_df[filtered_df["count"] >= 2]
```

```python
def check_floor(x):
    if 'floor' in x.lower():
        return True
    return False

filtered_df['contains_floor'] = filtered_df['Generation'].apply(check_floor)
filtered_df[filtered_df["contains_floor"] == True]
```

```python
import json
import re

def create_relational_datasets(input_file_path, output_on_path, output_under_path):
    """
    JSONL 파일을 읽어 'On' 관계와 'Under' 관계를 묻는 두 개의 데이터셋을 생성합니다.
    """
    
    # 질문 구조에서 (is/are), (obj1), (obj2)를 추출하기 위한 정규표현식
    # 예: Where is the beer bottle in relation to the armchair?
    # group(1): is/are
    # group(2): obj1 (smaller one)
    # group(3): obj2 (bigger one)
    pattern = r"Where (is|are) the (.+?) in relation to the (.+?)\?"
    
    on_dataset = []
    under_dataset = []
    
    try:
        with open(input_file_path, 'r', encoding='utf-8') as f:
            for line in f:
                entry = json.loads(line)
                
                question_text = entry.get('question', '')
                match = re.search(pattern, question_text)
                
                if match:
                    be_verb = match.group(1)      # is 또는 are
                    obj1 = match.group(2)         # 객체 1
                    obj2 = match.group(3)         # 객체 2
                    
                    # 원본 답변 추출 (대소문자 통일)
                    original_answer = entry['answer'][0].lower()
                    
                    # 문장 시작이므로 Be동사 첫 글자 대문자화
                    be_verb_cap = be_verb.capitalize()
                    
                    # -------------------------------------------------------
                    # 1. On Dataset 생성 로직
                    # -------------------------------------------------------
                    one_shot = "USER: Is the calculator on the desk? Answer with yes or no.\nASSISTANT: Think step by step. In this picture, the gray calculator is placed on top of the wooden desk, so the answer is yes.\n"
                    new_question_on = one_shot + f"<image>\nUSER: {be_verb_cap} the {obj1} on the {obj2}? Answer with yes or no.\nASSISTANT: Think step by step. "
                    
                    # 원본이 "on"일 때만 Yes, 나머지는 No
                    new_answer_on = "Yes" if original_answer == "on" else "No"
                    
                    on_dataset.append({
                        "id": entry['id'],
                        "question": new_question_on,
                        "answer": [new_answer_on]
                    })
                    
                    # -------------------------------------------------------
                    # 2. Under Dataset 생성 로직
                    # -------------------------------------------------------
                    new_question_under = f"<image>\nUSER: {be_verb_cap} the {obj1} under the {obj2}? Answer with yes or no.\nASSISTANT:"
                    
                    # 원본이 "under"일 때만 Yes, 나머지는 No
                    new_answer_under = "Yes" if original_answer == "under" else "No"
                    
                    under_dataset.append({
                        "id": entry['id'],
                        "question": new_question_under,
                        "answer": [new_answer_under]
                    })

        # 결과 저장
        with open(output_on_path, 'w', encoding='utf-8') as f_on:
            for item in on_dataset:
                f_on.write(json.dumps(item) + '\n')
                
        with open(output_under_path, 'w', encoding='utf-8') as f_under:
            for item in under_dataset:
                f_under.write(json.dumps(item) + '\n')
                
        print(f"성공적으로 처리되었습니다.")
        print(f"- On 데이터셋 저장됨: {output_on_path} (총 {len(on_dataset)}개)")
        print(f"- Under 데이터셋 저장됨: {output_under_path} (총 {len(under_dataset)}개)")

    except FileNotFoundError:
        print(f"에러: 입력 파일 '{input_file_path}'을 찾을 수 없습니다.")
    except Exception as e:
        print(f"처리 중 에러가 발생했습니다: {e}")

# 사용 예시
input_jsonl = "prompts/Controlled_Images_A_with_answer_four_options.jsonl"          # 원본 파일 경로
output_on_jsonl = "prompts/Controlled_Images_A_On_with_answer_two_options.jsonl"   # 결과 파일 경로 (On)
output_under_jsonl = "prompts/Controlled_Images_A_Under_with_answer_two_options.jsonl" # 결과 파일 경로 (Under)

create_relational_datasets(input_jsonl, output_on_jsonl, output_under_jsonl)
    
```

```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 노트북에서 그래프 바로 보이게
%matplotlib inline

# === 파일 경로 설정 ===
# 네 실험 경로에 맞게 바꿔줘
path_adapt = "./output_backup/Controlled_images_A/results1.5_Controlled_Images_A_adapt_vis_1.0_0.4_fouroption_True.json"
path_jsd   = "./output/results_Controlled_Images_A_adapt_vis_entropy_1.0_0.9_True.json"
```

```python
def load_results(path):
    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)
    return data

res_adapt = load_results(path_adapt)
res_jsd   = load_results(path_jsd)

print("len(adapt_vis) =", len(res_adapt))
print("len(adapt_vis_jsd) =", len(res_jsd))
```

```python
n = min(len(res_adapt), len(res_jsd))

rows = []
for i in range(n):
    a = res_adapt[i]
    j = res_jsd[i]

    # 안전하게 prompt / 정답이 진짜 같은지 확인
    if a["Golden"] != j["Golden"]:
        print(f"[WARN] Golden mismatch at index {i}: adapt={a['Golden']} jsd={j['Golden']}")
    # 필요하면 Prompt도 비교 가능
    # if a["Prompt"] != j["Prompt"]:
    #     print(f"[WARN] Prompt mismatch at index {i}")

    row = {
        "idx": i,
        "prompt": a.get("Prompt", None),
        "gold": a.get("Golden", None),

        "gen_adapt": a.get("Generation", None),
        "gen_jsd":   j.get("Generation", None),

        # 여기서 Uncertainty 필드 이름 맞추기
        "conf_adapt": a.get("uncertainty", None),  # max-softmax 기반
        "conf_jsd":   j.get("Uncertainty", None),  # JSD 기반
    }

    # 정답 여부 플래그 (네 코드랑 최대한 비슷하게)
    gold = str(row["gold"]).strip().lower() if row["gold"] is not None else None
    gen_a = (row["gen_adapt"] or "").strip().lower()
    gen_j = (row["gen_jsd"]   or "").strip().lower()

    correct_adapt = (gold is not None) and (gold in gen_a)
    correct_jsd   = (gold is not None) and (gold in gen_j)

    row["correct_adapt"] = correct_adapt
    row["correct_jsd"]   = correct_jsd

    rows.append(row)

df = pd.DataFrame(rows)
df.head()
```

```python
print("유효한 confidence 샘플 수 (둘 다 not None):",
      df[["conf_adapt", "conf_jsd"]].dropna().shape[0])

print("\n상관관계 (corr):")
print(df[["conf_adapt", "conf_jsd"]].corr())

print("\n정답 통계:")
print("adapt_vis 정확도:",
      df["correct_adapt"].mean() * 100)
print("adapt_vis_jsd 정확도:",
      df["correct_jsd"].mean() * 100)
```

```python
valid = df.dropna(subset=["conf_adapt", "conf_jsd"]).copy()

plt.figure(figsize=(6, 6))
plt.scatter(valid["conf_adapt"], valid["conf_jsd"], alpha=0.4, s=10)
plt.xlabel("max-softmax confidence (adapt_vis)")
plt.ylabel("entropy confidence (adapt_vis_entropy)")
plt.title("Relationship between max-softmax & entropy confidence")

x_vals = np.linspace(0, 1, 100) 
y_vals = -x_vals + 1

plt.plot(x_vals, y_vals, linestyle="--")
plt.grid(True)
plt.show()
```

```python
def categorize(row):
    if row["correct_adapt"] and row["correct_jsd"]:
        return "both_correct"
    elif (not row["correct_adapt"]) and (not row["correct_jsd"]):
        return "both_wrong"
    elif row["correct_adapt"] and (not row["correct_jsd"]):
        return "adapt_only"
    else:
        return "jsd_only"

valid["case"] = valid.apply(categorize, axis=1)

colors = {
    "both_correct": "tab:green",
    "both_wrong":   "tab:gray",
    "adapt_only":   "tab:blue",
    "jsd_only":     "tab:red",
}

plt.figure(figsize=(7, 7))
for case, sub in valid.groupby("case"):
    plt.scatter(
        sub["conf_adapt"],
        sub["conf_jsd"],
        alpha=0.6,
        s=18,
        label=case,
        color=colors.get(case, "black"),
    )

plt.xlabel("max-softmax confidence (adapt_vis)")
plt.ylabel("entropy confidence (adapt_vis_entropy)")
plt.title("Max-softmax vs Entropy by correctness case")
plt.legend()
plt.grid(True)
plt.show()
```
