```python
import pandas as pd
import json

r1_path = "output_backup/results_Controlled_Images_A_reasoning_1_1.0_1.0_False.json"
r2_path = "output_backup/results_Controlled_Images_A_reasoning_2_1.0_1.0_False.json"
r3_path = "output_backup/results_Controlled_Images_A_reasoning_3_1.0_1.0_False.json"

with open(r1_path, 'r') as f:
    r1_data = json.load(f)
    
with open(r2_path, 'r') as f:
    r2_data = json.load(f)
    
with open(r3_path, 'r') as f:
    r3_data = json.load(f)
    
r1_df = pd.DataFrame(r1_data)
r2_df = pd.DataFrame(r2_data)
r3_df = pd.DataFrame(r3_data)
```

```python
df = r1_df
valid_labels = ['Left', 'Right', 'Under', 'On']

correct_df = df[df['Generation'] == df['Golden']].copy()
wrong_df = df[df['Generation'] != df['Golden']].copy()  
correct_df.loc[~correct_df['Generation'].isin(valid_labels), 'Generation'] = 'Unknown'
wrong_df.loc[~wrong_df['Generation'].isin(valid_labels), 'Generation'] = 'Unknown'
target_index = valid_labels + ['Unknown']

correct_distribution = correct_df['Generation'].value_counts().reindex(target_index, fill_value=0)
wrong_distribution = wrong_df['Generation'].value_counts().reindex(target_index, fill_value=0)


print(f"Acc: {len(df[df['Generation'] == df['Golden']]) / len(df)}")
print('-'*30)
print("정답 케이스의 예측 분포:")
print(correct_distribution) 
print('-'*30)
print("오답 케이스의 예측 분포:")
print(wrong_distribution)
```
