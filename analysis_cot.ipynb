```python
import json, os
import pandas as pd
path = "output_backup/results1.5_Controlled_Images_A_few_shot_CoT_1.0_1.0_fouroption_True.json"

with open(path, 'r') as f:
    data = json.load(f)

df = pd.DataFrame(data)
df["last_sentence"] = df.apply(lambda row: row['Generation'].strip().split('.')[-2].strip(), axis=1)
df["is_correct_naive"] = df.apply(lambda row: row['Golden'].strip().lower() in row['Generation'].strip().lower(), axis=1)
df["is_correct_exact"] = df.apply(lambda row: row['Golden'].strip().lower() in row['last_sentence'].strip().lower(), axis=1)

naive = len(df[df["is_correct_naive"]==True]) / len(df)
exact = len(df[df["is_correct_exact"]==True]) / len(df)
print(f"Naive acc: {naive}")
print(f"Exact acc: {exact}")
```

```python
def filter_sentences_with_golden(row):
    # 1. Generation 텍스트 가져오기
    text = row['Generation']
    target = row['Golden']
    
    # 2. 마침표(.)를 기준으로 문장 분리
    # (빈 문장이 생길 수 있으므로 filtering도 함께 고려)
    sentences = [s.strip() for s in text.split('.') if s.strip()]
    
    # 3. Golden 라벨이 포함된 문장만 필터링 (대소문자 무시를 위해 lower() 사용)
    # 예: target="Right"라면 "right"가 포함된 문장도 찾음
    matched_sentences = [
        s for s in sentences 
        if target.lower() in s.lower()
    ]
    
    return matched_sentences

# apply 함수를 사용하여 새로운 컬럼 생성
df['extracted_sentences'] = df.apply(filter_sentences_with_golden, axis=1)

# 결과 확인
df[['Golden', 'extracted_sentences']]
```

```python

import json, os
import pandas as pd
path = "output/results_Controlled_Images_A_few_shot_CoT_1.0_1.0_True.json"


with open(path, 'r') as f:
    data = json.load(f)

df = pd.DataFrame(data)
```

```python
mask = df.apply(lambda row: str(row['Golden']).lower() in str(row['Answer']), axis=1)

# 2. 그룹 분리
group_included = df[mask].copy()      # 포함된 그룹 (정답)
group_not_included = df[~mask].copy() # 포함되지 않은 그룹 (오답)

# ---------------------------------------------------------
# 결과 확인
# ---------------------------------------------------------
print(f"전체 데이터 개수: {len(df)}")
print(f"포함된 그룹(Correct) 개수: {len(group_included)}")
print(f"포함되지 않은 그룹(Incorrect) 개수: {len(group_not_included)}")
```

```python
print(f"Acc: {100*(len(group_included) / len(df)):.04}%")
print(f"틀린 샘플 중.. Golden == 'Left': {100*(len(group_not_included[group_not_included['Golden'] == 'Left']) / len(group_not_included)):.04}%")
print(f"틀린 샘플 중.. Golden == 'Right': {100*(len(group_not_included[group_not_included['Golden'] == 'Right']) / len(group_not_included)):.04}%")
print(f"틀린 샘플 중.. Golden == 'On': {100*(len(group_not_included[group_not_included['Golden'] == 'On']) / len(group_not_included)):.04}%")
print(f"틀린 샘플 중.. Golden == 'Under': {100*(len(group_not_included[group_not_included['Golden'] == 'Under']) / len(group_not_included)):.04}%")
```
